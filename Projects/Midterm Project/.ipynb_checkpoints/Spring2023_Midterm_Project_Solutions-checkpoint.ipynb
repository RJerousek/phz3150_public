#### Part A. (30 pts) Create a function that considers the case of ZERO acceleration (constant velocity motion) in the $x$-direction and takes in 2 of the 3 kinematic variables: $\Delta x$, $v_x$,  and $t$ and calculates the remaining variable. Make sure to include an informative doc string and to test and account for errors based on the range of possible user inputs. Hint: You will need to use If/Else statements.

```python
def x_kin(delta_x = None, vx = None, t = None):
    
    """Considers the case of zero acceleration. Uses keywords so order of inputs doesn't matter. 
    Takes in inputs: delta_x = displacement in x direction in meters, vx = velocity in the x-direction in m/s, and
    t = the time in seconds. Determines whichever value is not given by the user. 
    Output: t, delta_x, v_x (in that order.)""" 
    
    if delta_x != None and vx != None and t != None:
        print("Problem is overconstrained!")
        return None
    
    if delta_x == None:
        if (vx == None) or (t == None):
            print("There are not enough given variables!")
            return None
        else:
            delta_x = vx*t
            
            
    if vx == None:
        if (delta_x == None) or (t == None):
            print("There are not enough given variables!")
            return None
        else:
            vx=delta_x/t
            
    if t == None:
        if (delta_x == None) or (vx == None):
            print("There are not enough given variables!")
            return None
        else:
            t = delta_x/vx
            
    return t,delta_x,vx
```

```python
print(x_kin(delta_x = 56.3,vx = 13.1))
```

#### Part B. (20 pts) Create a function that takes in the magnitude of a planet's constant acceleration due to gravity, $g$ and all but 2 of the 4 kinematic variables: $\Delta y$, $v_{0y}$, $v_{fy}$, and $t$ and calculates the 2 kinematic variables which were not given. Neglect air resistance. Make sure to include an informative doc string and to test and account for errors based on the range of possible user inputs. You can find a list of the four kinematic equations for constant acceleration in any introductory physics book.###

```python
def y_kin(g = None, delta_y = None, v_0 = None,v_f = None, t = None):
    
    """solves kinematic equations for constant acceleration for any two of the keyword inputs:
    delta_y = displacement in m, v_0 = initial velocity in m/s, v_f = final velocity in m/s, and t = time in seconds.
    the acceleration (g) is the (positive) magnitude of the accel. due to a planet's gravity at its surface. 
    output:all of the kinematic variables: g, delta_y, v_0, v_f, and t"""
    
    givens = [delta_y,v_0,v_f,t]
    nones = np.where(givens == None)
    if len(nones) > 2:
        print("There are not enough given variables.")
        return None
    if g == None:
        print,"Planetary surface gravity not given."
        return None
    
    g *= - 1 #acceleration should be downward.
    
    #it would be cleaner to use switchcase instead of if statements but this is how I expected you to solve it:
    
    if givens[0] == None:
        if givens[3] == None:    
            #1. Missing delta_y solve for t:
            t = (v_f - v_0)/g
        if givens[2] == None:
            #2. Missing delta_y solve for v_f:
            v_f = v_0 + g*t
        if givens[1] == None:
            #3. Missing delta_y solve for v_0:
            v_0 = v_f - g*t
    
    if givens[2] == None:
        if givens[3] == None:
            #4. Missing v_f solve for t:
            t1 = (-1*v_0 + np.sqrt(v_0**2 + 2*g*delta_y))/(2.*g)
            t2 = (-1*v_0 - np.sqrt(v_0**2 + 2*g*delta_y))/(2.*g)
            t = np.max((t1,t2))    #Take positive root (or really just max in this case)
        
        if givens[0] == None:
            #5. Missing v_f solve for delta_y:
            delta_y = v_0*t+0.5*g*t**2
            
        if givens[1] == None:
            #6. Missing v_f solve for v_0:
            v_0 = (delta_y - 0.5*g*t**2)/t
    
    if givens[3] == None:
        if givens[2] == None:
            #7. Missing t solve for v_f:
            v_f = np.sqrt(v_0**2 + 2*g*delta_y)#+/-
    
        if givens[1] == None:
            #8. Missing t solve for v_0:
            v_0 = np.sqrt(v_f**2 - 2*g*delta_y)
        
        if givens[0] == None:
            #9. Missing t solve for delta_y:
            delta_y = (v_f**2 - v_0**2)/2./g
            
    return g,delta_y,v_0,v_f,t
    
```

```python
print(y_kin(g = 9.8, delta_y = -10, v_0 = 3,v_f = None, t = None))
```

#### Part C. (10 pts) Write a function, kin_2D which calls one or both of the previous two functions and takes as input the (positive) magnitude of the acceleration due to gravity on a planet's surface, $g$, an initial position vector $\vec{r}_0 = (x_0,y_0)$ in m, a time $t$ in s, and the magnitude and angle (counterclockwise from the horizon or $+x$-axis) of the initial velocity $|\vec{v_0}|$ at angle $\theta$. The function should return the final position vector's $x$ and $y$ components in a 2 element array pos_f and the final velocity vector's x and y components in a 2 element array vel_f. Hint you must use trigonometry to break up the velocity vector into $x$ and $y$ components. 

```python
def kin_2d(g,pos_vec,t,v_0_mag,theta):
    
    """input: g (accel due to gravity in m/s/s), pos_vec (initial x and y components in m), t (time of flight in s),
    v_0_mag (magnitude of init. velocity), theta (angle of init velocity counterclockwise from horizon in deg.).
    output: final position and velocity vectors in x and y component form."""
    
    x0 = pos_vec[0]
    y0 = pos_vec[1]
    
    v0x = v_0_mag*np.cos(np.radians(theta))
    v0y = v_0_mag*np.sin(np.radians(theta))
    
    vfx = v0x #no acceleration in x direction.
    
    xf = v0x*t
    
    g, delta_y, v_0, v_f, t = y_kin(g = g, delta_y = None, v_0 = v0y,v_f = None, t = t)
    
    yf = y0 + delta_y
    
    pos_f = [xf,yf]
    vel_f = [vfx,v_f]
    
    return pos_f,vel_f


    
    
    
    
```

#### Part D. (10 points) Write a function make_plots, that takes in the same values as the previous problem but now also take in an array, $t$ of 100 times at 0.02 s intervals. The function should create 4 publication ready, labeled figures. Figure 1: The height of the projectile as a function of time $y(t)$. Figure 2: The magnitude of the velocity of the projectile $\vec{v}$ as a function of time. Figure 3: The height ($y$) of the projectile as a function of the $x$-position of the projectile (trajectory). Figure 4: The angle $\theta$ in which the projectile is moving in as a function of the projectile's $x$-position. (take the angle to be positive in the counterclockwise direction from the $+x$-axis and use units of degrees). Hint: Using a FOR loop for this is fine.

```python
def get_arrays(g,pos_vec,v_0_mag,theta):
    
    """Function that gets 2 by n element arrays for position and velocity
    Input: g = acceleration due to gravity in m/s/s, pos_vec = 2 element initial position (x0,y0) in m
    v0_mag = magnitude of initial velocity, theta = angle of launch in degrees counterclockwise from the +x-axis.
    Output: t = array of times, pos_f = positions (x and y), and vel_f = velocities (vx and vy)."""
    
    t = np.arange(100)*.02
    
    pos_f = np.zeros((2,len(t)))
    vel_f = np.zeros((2,len(t)))                   
    
    for i in range(len(t)):
        pos_f[:,i],vel_f[:,i] = kin_2d(g,pos_vec,t[i],v_0_mag,theta)
    
    return t, pos_f,vel_f

def make_plots(g,x0,y0,v_0_mag,theta):
    """Input: g = acceleration due to gravity in m/s/s, pos_vec = 2 element initial position (x0,y0) in m
    v0_mag = magnitude of initial velocity, theta = angle of launch in degrees counterclockwise from the +x-axis
    Output: none (creates for labeled figures.)"""
    
    r0 = [x0,y0]
    t, pos_f,vel_f = get_arrays(g,r0,v_0_mag,theta)
    
    fs = 20
    ymargin = 0.5

    #height plot:
    plt.figure(figsize = (12,4))
    plt.plot(pos_f[0,:],pos_f[1,:])
    plt.xlabel( 't'                 , fontsize = fs )
    plt.ylabel( 'y(t)' , fontsize = fs )
    plt.xticks( fontsize = fs )
    plt.yticks( fontsize = fs )
    plt.xlim(0,np.max(t))
    plt.ylim(0,np.max(pos_f[1,:]) + ymargin)
    plt.title("Figure 1", fontsize = fs)

    #velocity plot:
    plt.figure(figsize = (12,4))
    vmag = np.sqrt(vel_f[0,:]**2 + vel_f[1,:]**2)
    plt.plot(t,vmag)
    plt.xlabel( 't'                 , fontsize = fs )
    plt.ylabel( '|v(t)|' , fontsize = fs )
    plt.xticks( fontsize = fs )
    plt.yticks( fontsize = fs )
    plt.xlim(0,np.max(t))
    plt.ylim(0,np.max(vmag) + ymargin)
    plt.title("Figure 2", fontsize = fs)

    #trajectory plot:
    plt.figure(figsize = (12,4))
    plt.plot(pos_f[0,:],pos_f[1,:])
    plt.xlabel( 'x'                 , fontsize = fs )
    plt.ylabel( 'y' , fontsize = fs )
    plt.xticks( fontsize = fs )
    plt.yticks( fontsize = fs )
    plt.xlim(0,np.max(pos_f[0,:]))
    plt.ylim(np.min(pos_f[1,:]) - ymargin,np.max(pos_f[1,:]) + ymargin)
    plt.title("Figure 3", fontsize = fs)

    #angle plot:
    plt.figure(figsize = (12,4))
    theta = np.degrees(np.arctan(vel_f[1,:]/vel_f[0,:]))
    plt.plot(pos_f[0,:],theta)
    plt.xlabel( 'x'                 , fontsize = fs )
    plt.ylabel( '$\\theta$' , fontsize = fs )
    plt.xticks( fontsize = fs )
    plt.yticks( fontsize = fs )
    plt.xlim(0,np.max(pos_f[0,:]))
    plt.ylim(-90,90)
    plt.title("Figure 3", fontsize = fs)
                                               
    
    return None
    
```

```python
make_plots(9.8,0,5,4,70)
```

Kinetic Energy is $\frac{1}{2}mv^2$ and potential energy is $mgy$. The three curves should be in one figure using different colors and labeled with a legend showing the planet they took place on.

```python
g = [3.72,9.81,1.62]

fs = 15

print("1. Mars (g = 3.72m/s^2) v_0 = 20 m/s at 30 deg. x0 = 0 m y0 = 5 m") 
t_mars, p_mars, v_mars = get_arrays(3.72,[0,5],2,30)
print("2. Earth (g = 9.8m/s^2) v_0 = 7 m/s at -30 deg. x0 = 0 m y0 = 5 m") 
t_earth,p_earth,v_earth = get_arrays(9.8,[0,5],7,-30)
print("3. Moon (g = 1.62m/s^2) v_0 = 15 m/s at 10 deg. x0 = 3.3 m y0 = 10 m") 
t_moon, p_moon, v_moon = get_arrays(1.62,[3.3,10],15,10)



ymargin = 5.
E_mars = 0.5*(v_mars[0,:]**2 + v_mars[1,:]**2) + g[0]*p_mars[1,:]
E_earth = 0.5*(v_earth[0,:]**2 + v_earth[1,:]**2) + g[1]*p_earth[1,:]
E_moon = 0.5*(v_moon[0,:]**2 + v_moon[1,:]**2) + g[2]*p_moon[1,:]

plt.figure(figsize = (12,4))
plt.plot(t_mars,E_mars,color = "RED",label="Total Energy Mars")
plt.plot(t_earth,E_earth,color = "GREEN", label="Total Energy Earth")
plt.plot(t_moon,E_moon,color = "GREY", label="Total Energy Moon")
plt.xlabel( 't'                 , fontsize = fs )
plt.ylabel( 'Total Energy (J)' , fontsize = fs )
plt.xticks( fontsize = fs )
plt.yticks( fontsize = fs )
plt.xlim(0,np.max(t_mars))
plt.ylim(0,np.max(E_moon) + ymargin)
plt.legend(fontsize = fs)
```
